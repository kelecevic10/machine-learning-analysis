```python
import numpy as np
from pathlib import Path

data_path = Path("../../data/processed/preprocessed.npz")
data = np.load(data_path, allow_pickle = True)

X_train = data["X_train"]
y_train = data["y_train"]


print("X_train shape:", X_train.shape)
print("Y train shape:", y_train.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# uzimamo podskup zbog brzine
n_samples = 3000
idx = np.random.choice(len(X_train), n_samples, replace=False)

X_subset = X_train[idx]
y_subset = y_train[idx]

# PCA na 2 komponente
pca_2d = PCA(n_components = 2)
X_pca_2d = pca_2d.fit_transform(X_subset)

plt.figure(figsize=(8,6))

plt.scatter(
    X_pca_2d[:,0],
    X_pca_2d[:,1],
    c = y_subset,
    cmap = "tab20",
    s = 10
)

plt.title("PCA projekcija skupa podataka u 2D prostor")
plt.xlabel("Prva glavna komponenta (PC1)")
plt.ylabel("Druga glavna komponenta (PC2)")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# PCA na 3 komponente
pca_3d = PCA(n_components = 3)
X_pca_3d = pca_3d.fit_transform(X_subset)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(
    X_pca_3d[:,0],
    X_pca_3d[:,1],
    X_pca_3d[:,2],
    c = y_subset,
    cmap = "tab20",
    s = 10
)

ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")

plt.title("PCA projekcija skupa podataka u 3D prostor")

plt.show()
```
